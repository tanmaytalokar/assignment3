**1. What is the difference between a function and a method in Python?**

 Ans:- **Function**
       
  I. A function is a block of reusable code that can be defined using the def keyword.
      
  II. It is a general-purpose callable that can accept arguments and return values.
      
  III. A function is not bound to any object or class. It can be called directly.

  **Method:-**
      
  I.A method is a function that is associated with an object (usually an instance of a class).
      
  II.A method is always bound to the object it is called on, and it implicitly takes the object itself as the first argument, usually named self.
      
  III. Methods are invoked on objects using the dot notation.












```python
#EXAMPLE
#FUNCTION
def add(a, b):
    return a + b

result = add(3, 5)
print(result)

#METHOD
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(3, 5)
print(result)

```

**5. What are iterators in Python and how do they differ from iterables?**

Ans:-

  1. iterables and iterators are closely related but distinct concepts. An iterable is any object that can return an iterator. These objects are collections of data that can be looped over, such as lists, tuples, dictionaries, and sets. Essentially, an iterable is something you can pass to a for loop to iterate over its elements.

  2. To make an object iterable, it must implement the __iter__() method. This method returns an iterator object that is capable of keeping track of the iteration state and providing the next item in the sequence. You can think of iterables as data structures that organize and store data in a sequence.

  3. On the other hand, an iterator is an object that is responsible for actually iterating over the data. An iterator implements two important methods: __iter__() and __next__(). The __iter__() method returns the iterator object itself, and the __next__() method is used to get the next item in the sequence. When there are no more items to iterate over, __next__() raises a StopIteration exception to signal that the iteration is complete.




```python
#EXAMPLE
#ITERABLE
my_list = [1, 2, 3]
for item in my_list:
    print(item)

#ITERATOR
my_list = [4, 5, 6]
iterator = iter(my_list)

print(next(iterator))
print(next(iterator))
print(next(iterator))


```

**6. Explain the concept of generators in Python and how they are defined.**

 Ans:-

    1. A generator is a special type of iterator that allows you to iterate over a sequence of data lazily, meaning that it generates values on-the-fly as they are needed, rather than storing them all in memory at once.
    2. This is particularly useful when working with large datasets or sequences, as it reduces memory consumption by generating items one at a time instead of keeping everything in memory.
    3. Generators are defined using either a function with the yield keyword or using generator expressions. Let's look at both methods in detail.
    4. A generator is typically defined using a function, but instead of using a return statement to provide a result, you use the yield keyword. When a function contains a yield statement, it becomes a generator function. When you call this function, it does not run the function immediately; instead, it returns a generator object, which can be iterated over to get the values one by one.


```python
#EXAMPLE
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

counter = count_up_to(5)

for number in counter:
    print(number)
```

**7. What are the advantages of using generators over regular functions?**

Ans:-
    
  1. Regular functions typically generate and return all values at once, meaning they often require storing large collections of data in memory. This can be problematic when working with large datasets or sequences that cannot fit into memory.

  2. Generators support lazy evaluation, meaning that values are computed only when they are needed, not ahead of time. This is particularly useful in cases where the entire sequence may not need to be processed, or when the sequence is too large to process all at once.

```python
# EXAMPLE
def generate_numbers(n):
    for i in range(n):
        yield i

numbers = generate_numbers(100)
```

**9. Explain the purpose and usage of the `map()` function in Python.**

Ans:-  
 1. The map() function in Python is a built-in function that allows you to apply a given function to all items in an iterable (such as a list, tuple, or set) and return a map object (which is an iterator). This function is useful for applying transformations to data without needing to explicitly write loops, making the code more concise and readable.
 2. The map() function returns an iterator (a map object) that generates the transformed items one by one when iterated over. We can convert this iterator into other data structures like a list, tuple, or set if you need to use the results immediately.
 3. map() provides a more declarative way to apply a function to a collection of items, reducing the need for explicit loops. It can make the code cleaner and easier to read, especially when working with simple transformations.
 4. map() returns an iterator, it’s more memory efficient than using list comprehensions when dealing with large datasets, as the results are produced one by one, not stored all at once.

```python
# EXAMPLE
def square(x):
    return x ** 2
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

 Ans:-
   1. map() Function:
     
      I. The map() function applies a given function to every item in an iterable  and returns an iterator that produces the results.
      
      II. To apply a function to each item of the iterable(s) and transform the data.
      
      III. A function and one or more iterables.

    2. reduce() Function:
        
        I. The reduce() function is available in the functools module and applies a binary function cumulatively to the items of an iterable, from left to right, reducing the iterable to a single value.

        II. To reduce the iterable to a single value by applying a binary function cumulatively.
        
        III.A function (that takes two arguments) and an iterable.

     3. filter() Function:

        I. The filter() function filters an iterable by applying a function to each element and retaining only those elements where the function returns True.

        II. To filter the elements of an iterable based on a condition provided by a function.
        
        III. A function (that returns a boolean value) and an iterable.


```python
#EXAMPLE
#MAP FUNCTION
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))

#REDUCE FUNCTION
from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)

#FILTER FUNCTION
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))

```

```python
 #5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
 class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.current = 0
        self.next = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.current
        elif self.count == 1:
            self.count += 1
            return self.next
        else:
            fib = self.current + self.next
            self.current, self.next = self.next, fib
            self.count += 1
            return fib

if __name__ == "__main__":
    n_terms = 10
    fib_iterator = FibonacciIterator(n_terms)
    for fib in fib_iterator:
        print(fib, end=" ")
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

if __name__ == "__main__":
    max_exp = 10
    for power in powers_of_two(max_exp):
        print(power, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()

    except FileNotFoundError:
      print(f"Error: The file at '{file_path}' was not found.")
    except Exception as e:
      print(f"An error occured: {e}")
if __name__ == "__main__":
      file_path = "example.txt"
      for line in read_file_lines(file_path):
        print(line)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
celsius_temperatures = list(map(int, input("Enter temperatures in Celsius: ").split()))
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print("Temperatures in Celsius:", celsius_temperatures)
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```
